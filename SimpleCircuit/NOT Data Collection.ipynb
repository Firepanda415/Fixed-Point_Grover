```python
# Load IBMQ Account and choose a real backend
IBMQ.load_account()
provider = IBMQ.get_provider('ibm-q')
name = 'ibmqx2'
backend = provider.get_backend(name)
```

```python
# Global variable
n = 1
interested_qubits = [3,4] # Qubit Order
itrs = [1,3,5,10,15]

shots = 8192

DataFileAddress = 'NOTx2/'
NoiseFileAddress = 'NOTx2/'
# DataFileAddress = 'Bellbelem/'
# NoiseFileAddress = 'Bellbelem/'
```

```python
def counts_to_probs(counts):
    counts_shots = np.sum([v for v in counts.values()])
    probs = np.zeros(2**(n), dtype=np.float64)
    for k,v in counts.items():
        probs[int(k,2)] = v/counts_shots
    return probs

def VD(p,q):
    np_p = np.array(p)
    np_q = np.array(q)
    return 0.5*np.abs(np_p - np_q).sum()

def one_itr(qc):
    for i in range(10):
        qc.x(0)
        qc.barrier()


# Bell0 state
def pure_circuit(name=None):
    r = QuantumRegister(n)
    qc = QuantumCircuit(r, name=name)
    for it in range(itr):
        one_itr(qc)
    return r, qc  

def noise_gates_circuit(noise_params):
    angles = noise_params.reshape((3, n)) #NOT A NUMPY ARRAY FROM SCIPY.OPTIMIZE
    
    r = QuantumRegister(n)
    qc = QuantumCircuit(r)
    for it in range(itr):
        one_itr(qc)
        for q in range(n):
            qc.u(angles[0,n-q-1], angles[1,n-q-1], angles[2,n-q-1], n-q-1) # Add noise/denoise here
        qc.barrier()
    return r, qc
```

```python
# Generate Circuits and Record Them
circs = []
for q in interested_qubits:
    for itr in itrs:
        print("\n\nQubit {:d}, Number of Iterations = {:d}".format(q, itr))
        r, simple_qc = pure_circuit("NOTq{:d}Itr{:d}".format(q, itr))
        simple_state = Statevector.from_instruction(simple_qc)
        # Check state
        print("Ideal State", simple_state.data)
        # Measure
        simple_qc.measure_all()
        simple_qc.draw('mpl',filename=DataFileAddress+'NOT_circ_q{:d}_itr_{:d}_depth_{:d}.pdf'.format(q, itr, simple_qc.depth()))
        print("")
        # Transpile circuit
        trans_simple_qc = transpile(simple_qc, backend, initial_layout = [q], optimization_level=0)
        trans_simple_qc.draw('mpl',filename=DataFileAddress+'NOT_circ_trans_q{:d}_itr_{:d}_depth_{:d}.pdf'.format(q, itr, trans_simple_qc.depth()))
        print("Original Depth = {:d}, Transpiled Depth = {:d}".format(simple_qc.depth(), trans_simple_qc.depth()))
        circs.append(trans_simple_qc)
```

```python
# Submit to IBMQ
job_simple = execute(circs, backend=backend, shots=shots, memory=True, optimization_level=0)
job_monitor(job_simple)
```

```python
job_simple_id = job_simple.job_id()
np.savetxt(DataFileAddress+"Job_id_"+job_simple_id,np.array([]))
job_simple_id
```

```python
# Save Result
result_simple = job_simple.result()
counts = []
legends = []
for q in interested_qubits:
    for itr in itrs:
        mem_simple = np.array(result_simple.get_memory("NOTq{:d}Itr{:d}".format(q, itr)))
        np.save(DataFileAddress+'NOTq{:d}Itr{:d}.npy'.format(q, itr), mem_simple)

        counts.append(result_simple.get_counts("NOTq{:d}Itr{:d}".format(q, itr)))
        legends.append("Q{:d}Itr{:d}".format(q, itr))
```

```python
Q3_CB_mean = [0.91238178, 0.89138391, 0.0059039]
Q4_CB_mean = [0.88897331, 0.87090088, 0.00448789]
```

```python
from measfilter import getData0, errMitMat

def sxpower(s,x):
    total = 0
    length = len(s)
    for i in range(length):
        si = int(s[i])
        xi = int(x[i])
        total += si*xi
    return total

def count1(s):
    count = 0
    for i in range(len(s)):
        if s[i] == '1':
            count += 1
    return count

def gate_matrix(length, eps, m):
    size = 2**length
    mat = np.empty([size,size],dtype = np.float64)
    for row in range(size):
        for col in range(size):
            x = ("{0:0"+str(length)+"b}").format(row)
            s = ("{0:0"+str(length)+"b}").format(col)
            power = sxpower(s,x)
            mat[row,col] = ((-1)**power) * ((1 - eps)**(count1(s)*m))
    return mat

from cvxopt import matrix, solvers
def find_least_norm(ptilde): # only for one-qubit case
    # Formulation
    Q = 2*matrix(np.identity(2))
    p = -2*matrix(ptilde)

    G = matrix(np.array([[0,1],[0,-1]]),(2,2), 'd')
    h = 0.5*matrix(np.ones(2))

    A = matrix(np.array([1,0]),(1,2), 'd')
    b = matrix(0.5)

    solvers.options['show_progress'] = False
    sol=solvers.qp(Q, p, G, h, A, b)
    return sol['status'],sol['x']

def gate_denoise(p0s, lambdas):
    denoised = []
    meas_err_mat = errMitMat([lambdas[0], lambdas[1]])
    M = gate_matrix(1, lambdas[2], m)
    for p0 in p0s:
        ptilde = np.array([p0, 1 - p0])
        gate_ptilde = np.linalg.solve(meas_err_mat,ptilde)
        phat = np.linalg.solve(M,gate_ptilde)
        status, opt_phat = find_least_norm(phat)
        opt_recovered_p0 = opt_phat[0] + opt_phat[1] * (-1) ** (1 * 0) # phat(0) + phat(1)
        opt_recovered_p1 = opt_phat[0] + opt_phat[1] * (-1) ** (1 * 1) # phat(0) - phat(1)
        denoised.append(opt_recovered_p0)
        
    return denoised
```
